## Removing NULL from every df

```python
event_attendees_data.dropna(inplace=True)
user_friends_data.dropna(inplace=True)
users_df.dropna(inplace=True)
```

## Pre-Processing the event_attendees_data


```python
event_attendees_data
```

```python
for col in [x for x in event_attendees_data.columns if x != 'event']:
    event_attendees_data[col] = event_attendees_data[col].str.split()
```

```python
# new_events_attendees_data = pd.DataFrame(columns=['event', 'user'])
# for _, row in event_attendees_data.iterrows():
#     event_id = row['event']
#     user_list = row['yes']
#     for user_id in user_list:
#         new_events_attendees_data = pd.concat([new_events_attendees_data, pd.DataFrame({'event': [event_id], 'user': [user_id]})], ignore_index=True)
#     new_events_attendees_data
# new_events_attendees_data
new_events_attendees_data = event_attendees_data.explode('yes').rename(columns={'yes': 'user'})

# Rename the columns if needed
new_events_attendees_data = new_events_attendees_data[['event', 'user']]

new_events_attendees_data['went_to_event'] = 1
```

```python
new_events_attendees_data = pd.concat([new_events_attendees_data, event_attendees_data.explode('no').rename(columns={'no': 'user'})])

# Rename the columns if needed
new_events_attendees_data = new_events_attendees_data[['event', 'user', 'went_to_event']]

new_events_attendees_data.fillna(0, inplace=True)
```

```python
new_events_attendees_data
```

## Pre-Processing the user_friends

```python
user_friends_data['friends'] = user_friends_data['friends'].str.split()
```

```python
new_user_friends_data = user_friends_data.explode('friends')
```

```python
new_user_friends_data
```

```python
new_events_attendees_data
```

```python

merged_data = pd.merge(new_user_friends_data, new_events_attendees_data, left_on='friends',right_on='user', how='inner')
```

```python
merged_data.drop(columns=['user_y'], inplace=True)
merged_data.rename(columns={'user_x':'user'}, inplace=True)
```
