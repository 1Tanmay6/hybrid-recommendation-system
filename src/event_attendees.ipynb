```python
cols_to_convert = ['yes', 'maybe', 'invited', 'no']
for col in cols_to_convert:
    df[col] = df[col].str.split()
    df[col] = df[col].apply(lambda d: d if isinstance(d, list) else [])

df.head()
```

```python
def check_overlap(row):
    combined = row['yes'] + row['maybe'] + row['no']
    return len(combined) != len(set(combined))

# Apply the function to each row
# Apply the function to each row
df['overlap'] = df.apply(check_overlap, axis=1)

# Print 'event' and 'overlap' columns
df[['event', 'overlap']].head()

num_overlap = df['overlap'].sum()
print(num_overlap)

```

```python
def resolve_overlap(row):
    yes_set = set(row['yes'])
    maybe_set = set(row['maybe'])
    no_set = set(row['no'])
    
    # If data is overlapping between yes and no and maybe, pop yes and no
    overlap_yes_no_maybe = yes_set & no_set & maybe_set
    row['yes'] = list(yes_set - overlap_yes_no_maybe)
    row['no'] = list(no_set - overlap_yes_no_maybe)

    # If data is overlapping between yes and no, pop both and put it in maybe
    overlap_yes_no = yes_set & no_set
    row['yes'] = list(yes_set - overlap_yes_no)
    row['no'] = list(no_set - overlap_yes_no)
    row['maybe'] += list(overlap_yes_no)

    # If data is overlapping between yes and maybe, pop the maybe entry
    overlap_yes_maybe = yes_set & maybe_set
    row['maybe'] = list(maybe_set - overlap_yes_maybe)

    # If data is overlapping between no and maybe, pop the no entry
    overlap_no_maybe = no_set & maybe_set
    row['no'] = list(no_set - overlap_no_maybe)


    return row

# Apply the function to each row
df = df.apply(resolve_overlap, axis=1)
df['overlap'] = df.apply(check_overlap, axis=1)

# Print 'event' and 'overlap' columns
print(df[['event', 'overlap']])

num_overlap = df['overlap'].sum()
print(num_overlap)
```

```python
# Initialize an empty list
data = []

for idx, row in df.iterrows():
    invited_set = set(row['invited'])
    for user in row['yes']:
        data.append({'event': row['event'], 'user': user, 'yes': 1, 'maybe': 0, 'no': 0, 'invited': 1 if user in invited_set else 0})
    for user in row['maybe']:
        data.append({'event': row['event'], 'user': user, 'yes': 0, 'maybe': 1, 'no': 0, 'invited': 1 if user in invited_set else 0})
    for user in row['no']:
        data.append({'event': row['event'], 'user': user, 'yes': 0, 'maybe': 0, 'no': 1, 'invited': 1 if user in invited_set else 0})

# Convert the list to a DataFrame
new_df = pd.DataFrame(data)

new_df.head(15)
```

```python
def calculate_score(row):
    if row['invited']:
        if row['yes']:
            return 0.9
        elif row['maybe']:
            return 0.5
        elif row['no']:
            return 0.1
    else:
        if row['yes']:
            return 1
        elif row['maybe']:
            return 0.5
        elif row['no']:
            return 0

new_df['weighted_score'] = new_df.apply(calculate_score, axis=1)

new_df.head(15)
```
